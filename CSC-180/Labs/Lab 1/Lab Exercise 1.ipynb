### **1. Reading CSV File into Dataframe**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("./homeprices.csv")
df.head()
```

### **2. Training/Fitting Model**

```python
model = linear_model.LinearRegression()
model.fit(df[['area']], df.price)
```

### **4. Saving Model (pickle)**

```python
import pickle
```

```python
with open('model_pickle', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('model_pickle', 'rb') as f:
    mp = pickle.load(f)
```

Using pickle module we can create memory that holds the model. Pickle is the built in standard library in Python for this. It works, but can be slow for large data sets.

### **5. Saving Model (joblib)**

```python
import joblib
```

```python
joblib.dump(model, 'model_joblib')
```

```python
mj = joblib.load('model_joblib')
```

Joblib is an external module outside of the standard Python libraries that essentially does the same thing as pickle but in simple syntax. Joblib also excels better than pickle for larger data sets, I believe most times we will use this as a method of saving the model.

```python
mj.predict([[5000]])
```
