### **1. Reading CSV File into Dataframe**

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('./homePricesMedian.csv')
df
```

Notice that some values in the dataframe are not available, for whatever the reason. Values must be numerical throughout the entire column in order for the model to accept the input. We have to substitute something in somehow.

### **2. Filling NaN Entries (Data pre-processing)**

```python
df.bedrooms.median()
```

The most logical substitutions for missing values seem to be either the median value of the remaining existing ones or the mean value. This is a developer decision and we should pick whichever one matches the data most closely.

```python
import math
median_bedrooms = math.floor(df.bedrooms.median()) # Take the floor, only integers allowed. can't have half a bedroom.
median_bedrooms
```

```python
df.bedrooms.fillna(median_bedrooms)
```

```python
df.bedrooms = df.bedrooms.fillna(median_bedrooms)
df
```

### **3. Training/Fitting Model**

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area', 'bedrooms', 'age']], df.price)
```

Creates the model based off the inputs: area, bedrooms, age, and outputs: price.

```python
reg.coef_
```

```python
reg.intercept_
```

### **4. Test Cases (Predictions)**

#### **Test Case 1**

```python
reg.predict([[3000, 3, 40]])
```

```python
137.25*3000+-26025*3+-6825*40+383724.9999999983
```
