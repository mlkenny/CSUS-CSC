```python
import tkinter as tk
import joblib
from tkinter import ttk

def calculate_price():
    try:
        model = car_model.get()
        age = int(age_entry.get())
        mileage = int(mileage_entry.get())

        '''This section is for a linear model based on the base prices of models.'''
        '''This does not reflect anything about the model we create with the given data.'''
        # # Base prices depending on model
        # base_prices = {
        #     "BMW": 50000,
        #     "Audi": 45000,
        #     "Mercedes": 55000
        # }

        # base_price = base_prices.get(model, 30000)

        # # Formula: price decreases with age and mileage
        # price = base_price - (age * 1000) - (mileage * 0.05)

        # if price < 1000:
        #     price = 1000  # minimum price

        '''This sections uses either the linear regression model using dummies,'''
        '''or it uses the linear regression model using ohe, according to the selected model.'''
        # Model functions off inputs: [[Mileage, Age, Audi, BMW]]
        # Selected model should represent the dummies model or the ohe model, based on radio button.

        match select_model():
            case 1:
                selected_model = joblib.load('model_dummies')
            case 2:
                selected_model = joblib.load('model_ohe')

        match model:
            case "BMW":
                price = selected_model.predict([[mileage, age, 0, 1]])
            case "Audi":
                price = selected_model.predict([[mileage, age, 1, 0]])
            case "Mercedes":
                price = selected_model.predict([[mileage, age, 0, 0]])

        result_label.config(text=f"Estimated Price: ${price[0]:,.2f}")
    except ValueError:
        result_label.config(text="Please enter valid numbers for age and mileage.")

def reset_fields():
    car_model.set("")
    age_entry.delete(0, tk.END)
    mileage_entry.delete(0, tk.END)
    selection.set(1) # Added reset model to dummies if fields reset.
    result_label.config(text="Estimated Price: ")
    
def select_model():
    return selection.get()

# GUI window
root = tk.Tk()
root.title("Car Price Estimator")
root.geometry("400x300")

# Car Model
tk.Label(root, text="Car Model:").pack()
car_model = ttk.Combobox(root, values=["BMW", "Audi", "Mercedes"])
car_model.pack()

# Age
tk.Label(root, text="Age (years):").pack()
age_entry = tk.Entry(root)
age_entry.pack()

# Mileage
tk.Label(root, text="Mileage (km):").pack()
mileage_entry = tk.Entry(root)
mileage_entry.pack()

# Buttons
button_frame = tk.Frame(root)
button_frame.pack(pady=10)

calc_button = tk.Button(button_frame, text="Calculate Price", command=calculate_price)
calc_button.grid(row=0, column=0, padx=5)

reset_button = tk.Button(button_frame, text="Reset", command=reset_fields)
reset_button.grid(row=0, column=1, padx=5)

# Radio Button
selection = tk.IntVar() # We need a variable to store the value of the radio button.
selection.set(1) # By default selection should be set to 1, this defaults the model to dummies.

dummies_button = tk.Radiobutton(root, text="Dummies", variable=selection, value=1)
dummies_button.pack(pady=1)

ohe_button = tk.Radiobutton(root, text="OHE", variable=selection, value=2)
ohe_button.pack(pady=1)

# Result Label
result_label = tk.Label(root, text="Estimated Price: ")
result_label.pack(pady=10)

root.mainloop()
```
